### b) d = $\frac{\begin{bmatrix} 4 \\ -0.5 \end{bmatrix}^T * \begin{bmatrix} 3 \\ 4 \end{bmatrix} + 5}{||\begin{bmatrix} 3 \\ 4 \end{bmatrix}||}$ = 3

```python
import numpy as np
def signed_dist(x, theta, theta_0):
    top = np.dot(x.T, theta)[0][0] + theta_0
    bottom = np.linalg.norm(theta)
    return top / bottom

p = np.array([
    [-2],
    [3]
])

th = np.array([
    [1],
    [-1]
])

signed_dist(p, th, 4)
```

### Question 3

### a) If the units were changed from meters to millimeters, the value of R would increase since each meter is 1000 millimeters. However, the size of the margin would also grow, since the hyperplane would not change (data scale does not affect the hyperplane), which would lead to the upper bound of guesses remaining the same.

```python
def mistake_bound(R, gamma):
    return (R/gamma) ** 2


for g in [0.001, 0.01, 0.1, 0.2, 0.5]:
    print(f"gamma = {g}, error bound : {mistake_bound(1, g)}")
```
